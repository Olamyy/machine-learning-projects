```python
estimated_deaths_dataframe = estimated_deaths_dataframe.drop(['YEAR (DISPLAY)'], axis=1)
```

```python
estimated_deaths_dataframe.head(5)
```

```python
estimated_deaths_dataframe.columns = ['PUBLISH STATE', 'YEAR ', 'REGION','COUNTRY', 'DISPLAY VALUE','NUMERIC', 'LOW', 'HIGH']
```

```python
estimated_deaths_dataframe.head(5)
```

```python
estimated_deaths_dataframe.isnull().sum()
```

```python
africa = estimated_deaths_dataframe[estimated_deaths_dataframe['REGION'] == 'Africa']
americas = estimated_deaths_dataframe[estimated_deaths_dataframe['REGION'] == 'Americas']
europe = estimated_deaths_dataframe[estimated_deaths_dataframe['REGION'] == 'Europe']
east_med = estimated_deaths_dataframe[estimated_deaths_dataframe['REGION'] == 'Eastern Mediterranean']
west_pacific = estimated_deaths_dataframe[estimated_deaths_dataframe['REGION'] == 'Western Pacific']
south_east_asia = estimated_deaths_dataframe[estimated_deaths_dataframe['REGION'] == 'South-East Asia']

regions = [africa, americas, europe, east_med, west_pacific,south_east_asia]

africa_sum = africa.NUMERIC.sum()
americas_sum = americas.NUMERIC.sum()
europe_sum = europe.NUMERIC.sum()
east_med_sum = east_med.NUMERIC.sum()
west_pacific_sum = west_pacific.NUMERIC.sum()
south_east_asia_sum = south_east_asia.NUMERIC.sum()
```

```python
africa.head(5)
```

```python
sums = [africa_sum, europe_sum, americas_sum, east_med_sum, west_pacific_sum, south_east_asia_sum]
sum_barrier = min(sums)
```
